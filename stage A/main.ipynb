```python
import numpy as np
import pandas as pd
```

```python
# Reading the csv 
data = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding = 'latin-1')
```

```python
data.head()
```

```python
# What is the total number and percentage of missing data in 2014 to 3 decimal places?

total_missing = data['Y2014'].isnull().sum()
percentage_missing = (total_missing / len(data['Y2014'])) * 100

print(f'The total number of missing data is {total_missing} and the percentage of missing data is {percentage_missing:.3f}%')
```

```python
# Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?
grp1 = data.groupby('Item')
 
grp1.aggregate(np.sum)
```

```python
# Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?

# Groupby operation on 'Element'

data.groupby('Element').sum().sort_values(by = 'Y2017', ascending = False)
```

```python
# What is the total number of unique countries in the dataset?

# Total number of unique countries in the dataset

unique_countries = data['Area'].nunique()

unique_countries
```

```python
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

# Selecting columns 'Y2017' and 'Area'

data[['Y2017', 'Area']]
data.groupby('Area').sum().sort_values(by = 'Y2017', ascending = False).head(1)
```

```python
# Which year had the least correlation with ‘Element Code’?

# Year with the least correlation with 'Element Code'

data.corr().loc['Element Code']
```

```python
# What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

# Mean and standard deviation across the whole dataset for the year 2017

mean_2017 = data['Y2017'].mean()
std_2017 = data['Y2017'].std()

print(f'The mean for the year 2017 is {mean_2017:.2f} and the standard deviation is {std_2017:.2f}')
```

```python
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

# Selecting columns 'Y2017' and 'Area'

data[['Y2017', 'Area']]
data.groupby('Area').sum().sort_values(by = 'Y2017', ascending = True).head(7)
```

```python
# Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

# Groupby operation on 'Element'

data.groupby('Element').sum().sort_values(by = 'Y2017', ascending = False)
```
